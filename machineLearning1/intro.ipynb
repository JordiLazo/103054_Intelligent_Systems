# Our first model

```python
import sklearn.linear_model

X = diabetes.data[["bmi"]]
y = diabetes.target

model = sklearn.linear_model.LinearRegression()

model.fit(X, y)
```

Let's see the parameters of this model and its values

```python
print(model.coef_)
print(model.intercept_)

def calculate(example):
    # y = c0 + c1 * f1 + c2 * f2...
    _y = model.intercept_
    for coef, feat_val in zip(model.coef_, example):
        print("##", coef, feat_val)
        _y += coef * feat_val
    return _y 

example = [10]
print(calculate(example))
print(model.predict([example]))
```

```python
y_pred = model.predict(X)

seaborn.scatterplot(x=X.bmi, y=y)
seaborn.lineplot(x=X.bmi, y=y_pred, color="red")
```
